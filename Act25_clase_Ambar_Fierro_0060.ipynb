### Crear la clase Película, con 3 atributos (título, género, duración), 2 funciones leerdatos(), mostrardatos() y 2 objetos (pelicula1, pelicula2).

![texto del vínculo](https://raw.githubusercontent.com/a25308051280060-pixel/Act25-Clase-pel-cula-Ambar-fierro-0060/refs/heads/main/infoclasepelicula.png.png)

```python
class Pelicula:
    def __init__(self, titulo, genero, duracion):
        self.titulo = titulo
        self.genero = genero
        self.duracion = duracion

    def leerdatos(self):
        # Esta función simula la lectura o confirmación de datos del objeto
        print(f"Cargando datos de la película: {self.titulo}...")

    def mostrardatos(self):
        print(f"Película: {self.titulo} | Género: {self.genero} | Duración: {self.duracion}")

# --- Instanciación (2 objetos) ---
pelicula1 = Pelicula("Inception", "Ciencia Ficción", "2h 28min")
pelicula2 = Pelicula("El Padrino", "Drama/Crimen", "2h 55min")

# --- Llamadas a las funciones para cada objeto ---
print("--- Info Película 1 ---")
pelicula1.leerdatos()
pelicula1.mostrardatos()

print("\n--- Info Película 2 ---")
pelicula2.leerdatos()
pelicula2.mostrardatos()

print("\n" + "-"*30)
print("creador de software por Ambar Fierro NL = 19")
print("programa terminado")
```
